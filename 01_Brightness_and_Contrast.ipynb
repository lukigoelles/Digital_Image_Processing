# Brightness and contrast adjustment

```python
from PIL import Image as im
import numpy as np

image = im.open("demo.jpg")
display(image)
```

We change the brightness of an image by multiplying each pixel by a linear gain parameter $g$, 
\begin{align}
    f_2[x,y] = g \cdot f_1[x,y] \, .
\end{align}

```python
im_matrix = np.array(image)
gain = 3
im_matrix2 = gain * np.array(im_matrix, dtype=np.uint32)

for ii in range(image.height):
    for jj in range(image.width):
        maximum = np.max(im_matrix2[ii,jj])
        if maximum > 255:
            im_matrix2[ii,jj] = im_matrix2[ii,jj]/maximum*255

image2 = im.fromarray(np.array(im_matrix2, dtype=np.uint8))
display(image2)
```

The contrast is change by introducing an exponent $\gamma$,
\begin{align}
    f_2[x,y] = f_1[x,y] ^ \gamma \, .
\end{align}

```python
gamma = 1.2
im_matrix3 = np.array(im_matrix, dtype=np.uint32)**gamma

for ii in range(image.height):
    for jj in range(image.width):
        maximum = np.max(im_matrix3[ii,jj])
        if maximum > 255:
            im_matrix3[ii,jj] = im_matrix3[ii,jj]/maximum*255

image3 = im.fromarray(np.array(im_matrix3, dtype=np.uint8))
display(image3)
```
